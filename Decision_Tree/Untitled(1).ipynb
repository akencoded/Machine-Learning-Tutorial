```python
import pandas as pd
```

```python
dataset = pd.read_csv(r'C:\Users\abhi2\Downloads\archive (1)\tennis.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder   #converting given data into numbers
```

```python
outlook=LabelEncoder()
temp=LabelEncoder()
humidity=LabelEncoder()
windy=LabelEncoder()
play=LabelEncoder()
```

```python
dataset['outlook']=outlook.fit_transform(dataset['outlook'])   #replacing labels into numbers
dataset['temp'] = outlook.fit_transform(dataset['temp'])
dataset['humidity'] = outlook.fit_transform(dataset['humidity'])
dataset['windy'] = outlook.fit_transform(dataset['windy'])
dataset['play'] = outlook.fit_transform(dataset['play'])
```

```python
dataset
```

# separating IV AND DV

```python
features_cols = ['outlook','temp','humidity','windy']
X=dataset[features_cols]
y=dataset.play
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test, = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#classifier = DecisionTreeClassifier(criterion='gini')
classifier = DecisionTreeClassifier(criterion='entropy')
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
X_test     #0->no  1->yes  sunny->2  overcast->0  rainy->1
```

```python

classifier.score(X_test,y_test)
```
